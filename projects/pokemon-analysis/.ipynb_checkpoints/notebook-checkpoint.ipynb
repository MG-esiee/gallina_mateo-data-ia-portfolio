# Pokémon Marketing Analysis

Ce notebook analyse un dataset Pokémon afin d'identifier des tendances et des groupes marketing. 

Objectifs :
- Comprendre les caractéristiques des Pokémon
- Segmenter les Pokémon en clusters basés sur leurs stats
- Générer des insights visuels et analytiques pour des applications mar

## Import des librairies

Nous importons les librairies nécessaires pour :
- la manipulation des données (pandas, numpy)  
- la visualisation (matplotlib, seaborn)  
- le clustering (scikit-learn)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

#couleru pour graphiques seaborn
sns.set(style="whitegrid")
```

## Chargement et exploration des données

Nous chargeons le dataset Pokémon et inspectons les premières lignes, le type des colonnes, et la présence de valeurs manquantes.

```python
df = pd.read_csv("data/pokemon.csv")

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

## Nettoyage des données

- On supprime ou remplit les valeurs manquantes si nécessaire.
- On conserve uniquement les colonnes utiles pour l'analyse marketing :
  'Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total'


```python
df = df[['Name','Type 1','Type 2','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Total']]

df['Type 2'] = df['Type 2'].fillna('None')

df.head()
```

## Visualisation des données

Nous explorons la distribution des statistiques et des types de Pokémon pour comprendre les tendances.


```python
#histo des stats
stats = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Total']
df[stats].hist(figsize=(12,8))
plt.suptitle("Distribution des statistiques des Pokémon")
plt.show()
```

```python
#countplot des types principaux
plt.figure(figsize=(12,6))
sns.countplot(x='Type 1', data=df, order=df['Type 1'].value_counts().index)
plt.title("Distribution des Type 1")
plt.xticks(rotation=45)
plt.show()
```

## Feature Engineering

- On normalise les statistiques pour le clustering.
- On prépare les features numériques pour KMeans.

```python
features = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Total']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Clustering

Nous utilisons KMeans pour segmenter les Pokémon en 5 clusters.  
Chaque cluster représente un groupe de Pokémon aux caractéristiques similaires.


```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

#visualiser la distrib des clusters
sns.countplot(x='Cluster', data=df)
plt.title("Nombre de Pokémon par cluster")
plt.show()
```

## Analyse des clusters

Nous examinons les caractéristiques de chaque cluster et identifions les traits dominants.


```python
#moyenne des stats par cluster
cluster_summary = df.groupby('Cluster')[features].mean()
cluster_summary
```

```python
#visualisation des clusters sur Total vs Attack
plt.figure(figsize=(10,6))
sns.scatterplot(x='Attack', y='Total', hue='Cluster', data=df, palette='tab10')
plt.title("Clusters des Pokémon : Total vs Attack")
plt.show()
```

## Insights

- Chaque cluster regroupe des Pokémon avec des caractéristiques similaires.
- On peut identifier les clusters avec les Pokémon les plus forts, les plus rapides, ou les plus défensifs.
- Ces insights peuvent être utilisés pour le marketing, le jeu, ou la segmentation de produits dérivés.

Le notebook montre l'ensemble du workflow : exploration, nettoyage, visualisation, feature engineering, clustering et analyse finale.
